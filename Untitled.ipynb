```python
import pandas as pd
```

```python
#Read the Data File
df=pd.read_csv('car data.csv')
```

```python
#Display the Features of Dataset
df.head()
```

```python
df.shape
```

```python
df.describe()
```

```python
#Display unique values of categorical features
print(df['Seller_Type'].unique())
print(df['Fuel_Type'].unique())
print(df['Transmission'].unique())
print(df['Owner'].unique())
```

```python
#Check Missing Values
df.isnull().sum()
```

```python
#Create a derived feature of Number of Years
df['Current_Year']=2020
df.head()
```

```python
df['Number of Years']=df['Current_Year']-df['Year']
df.head()
```

```python
df.columns
```

```python
#Dropping the not required columns
finaldataset=df[['Selling_Price', 'Present_Price',
       'Kms_Driven', 'Fuel_Type', 'Seller_Type', 'Transmission', 'Owner','Number of Years']]
finaldataset.head()
```

```python
#Get Dummy Variables for Categorical Dataset
finaldataset=pd.get_dummies(finaldataset,drop_first=True)
finaldataset.head()
```

```python
#Show Correlations
finaldataset.corr()
```

```python
#Display Correlations Graphically
import seaborn as sns
sns.pairplot(finaldataset)
```

```python
#Display Heat Map
import matplotlib.pyplot as plt
corrmat=finaldataset.corr()
top_corr_features=corrmat.index
plt.figure(figsize=(20,20))
#plot heat map
g=sns.heatmap(finaldataset[top_corr_features].corr(),annot=True,cmap="RdYlGn")
```

```python
# Separate Dependent and Independent Features
```

```python
X=finaldataset.iloc[:,1:]
Y=finaldataset.iloc[:,0]
```

```python
X.head()
```

```python
Y.head()
```

```python
# Find the Feature Importance
from sklearn.ensemble import ExtraTreesRegressor
model=ExtraTreesRegressor()
model.fit(X,Y)
print(model.feature_importances_)
```

```python
#Plot Feature Importances
feat_importances=pd.Series(model.feature_importances_,index=X.columns)
feat_importances.nlargest(5).plot(kind='barh')
plt.show()
```

```python
# Train Test Split
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2)
```

```python
X_train.shape
```

```python
# Implement Random Forest Regressor
from sklearn.ensemble import RandomForestRegressor
rf_random=RandomForestRegressor()
```

```python
# Hyperparameters selection
import numpy as np
n_estimators=[int(x) for x in np.linspace (start=100,stop=1200,num=12)] #Creating number of decision tress
print(n_estimators)
```

```python
#Number of Trees in Random Forest
n_estimators=[int(x) for x in np.linspace (start=100,stop=1200,num=12)]

#Number of Features to consider at every split
max_features=['auto','sqrt']

#Maximum number of levels in Tree
max_depth=[int(x) for x in np.linspace (5,30,num=60)]

#max_depth.append(none)
#Minimum number of samples required to split a node
min_samples_split=[2,5,10,15,100]

# Minimum number of samples required at each leaf node
min_samples_leaf=[1,2,5,10]
```

```python
#Finding best combination of Parameters from above combinations
from sklearn.model_selection import RandomizedSearchCV
random_grid={'n_estimators':n_estimators,
            'max_features':max_features,
            'max_depth':max_depth,
            'min_samples_split':min_samples_split,
            'min_samples_leaf':min_samples_leaf}
print(random_grid)
```

```python
#Use Random Grid to search for best hyperparameters
#Create a base mode
rf=RandomForestRegressor()
rf_random=RandomizedSearchCV(estimator=rf,param_distributions=random_grid,scoring='neg_mean_squared_error',n_iter=10,cv=5,verbose=2,random_state=42,n_jobs=1)
```

```python
rf_random.fit(X_train,Y_train)
```

```python
#Make the Predictions on Test Dataset
predictions=rf_random.predict(X_test)
predictions
```

```python
#Compare the Y_test and Predictions
sns.distplot(Y_test-predictions)
```

```python
plt.scatter(Y_test,predictions)
```

```python
#Requirement for Deployment
import pickle
#open a file where you want to store the data
file=open('random_forest_regression_model.pkl','wb')

#dump information to the file
pickle.dump(rf_random,file)
```
